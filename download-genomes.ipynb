```python
import pandas as pd
import ftputil
import gzip
import re
import os
```

```python
viruses = pd.read_csv("viruses2.csv")
viruses.head()
```

```python
with open('totalVirusDna.txt', 'w') as fout:
    for index in range(0,viruses.shape[0]):
        if type(viruses.iloc[index]['GenBank FTP']) == type(""):
            ftpFolder = viruses.iloc[index]['GenBank FTP'][26:]
            #print(ftpFolder)
            ftpFile = viruses.iloc[index]['Assembly']+'_ViralProj14066_genomic.fna.gz';
            #print(ftpFile)
            host = ftputil.FTPHost('ftp.ncbi.nlm.nih.gov', 'anonymous', 'password')
            host.chdir(ftpFolder)
            allFiles = host.listdir(host.curdir)
            genomeFile = list(filter(lambda x: '.fna.gz' in x,allFiles))
            if(len(genomeFile) >= 1):
                host.download(genomeFile[0],genomeFile[0])
                with gzip.open(genomeFile[0], 'rb') as f:
                    rawSplits = f.read().decode('UTF-8').split('\n')
                    dnaOnly = list(filter(lambda x: re.fullmatch('^[ACTG]*',x),rawSplits))
                    fullDnaString = ''.join(dnaOnly)
                    fout.write(fullDnaString)
                    fout.write('\n')
                os.remove(genomeFile[0])
        print(str(index) + ' ' + str(100*(index/viruses.shape[0]) )+'%')
```
