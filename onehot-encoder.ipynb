```python
from IPython.display import clear_output
import numpy as np
f = open("totalVirusDna.txt", "r")
lines = f.readlines()
count=0
onehotList=[]
for line in lines:
    if len(line) <= 25000:
        seq = []
        for ci in range(0,25001):
            if ci < len(line):
                if line[ci] == 'A':
                    seq.append([1,0,0,0,0])
                elif line[ci] == 'T':
                    seq.append([0,1,0,0,0])
                elif line[ci] == 'C':
                    seq.append([0,0,1,0,0])
                elif line[ci] == 'G':
                    seq.append([0,0,0,1,0])
            else:
                seq.append([0,0,0,0,1])
        onehotList.append(seq)
    #print(str(100*count/len(lines)))

print(len(onehotList))
print(lines[0])
```

```python
onehotArray = np.array(onehotList)
print(onehotArray.shape)
np.save('oneHotDna.npy',onehotArray)
```
